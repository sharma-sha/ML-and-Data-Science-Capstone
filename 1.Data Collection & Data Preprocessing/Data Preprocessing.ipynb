```python
for column in ['job_category', 'salary_currency', 'experience_level', 'company_size']: dataset[column].fillna(dataset[column].mode()[0], inplace=True)
```

```python
qual=[]
quan=[]
for columnName in dataset.columns:
    print(columnName)
    if(dataset[columnName].dtype=='O'):
        #print("qual")
        qual.append(columnName)
    else:
        #print("quan")
        quan.append(columnName)
```

```python
qual
```

```python
quan
```

```python
#fn
def quanQual(dataset):
    qual=[]
    quan=[]
    for columnName in dataset.columns:
        print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual        
            
```

```python
quanQual(dataset)
```

```python
quan,qual=quanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
#https://scikit-learn.org/stable/modules/impute.html
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df=pd.DataFrame(df,columns=quan)
```

```python
df
```

```python
cate=dataset[qual]
```

```python
cate
```

```python
two=[df,cate]
```

```python
preprocessed=pd.concat(two,axis=1)
```
