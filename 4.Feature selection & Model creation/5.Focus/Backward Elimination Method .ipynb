```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv("DS_jOBs.csv")
```

```python
df=pd.get_dummies(df,dtype=int,drop_first=True)
```

```python
df
```

```python
print(df.dtypes)
```

```python
X = df.iloc[:, 0:45]
y = df.iloc[:, -1]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
from sklearn.ensemble import RandomForestClassifier
```

```python
bakward_feature_selector = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                    k_features=5,
                                                    forward=False,
                                                    floating=False,
                                                    verbose=2,
                                                    scoring= "accuracy",
                                                    cv= 5).fit(X_train, y_train)
```

```python
bakward_feature_selector.k_feature_idx_
```

```python
bakward_feature_selector.k_feature_names_
```

```python
bakward_feature_selector.k_score_
```
