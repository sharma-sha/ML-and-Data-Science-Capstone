```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv("DS_jOBs.csv")
```

```python
df=pd.get_dummies(df,dtype=int,drop_first=True)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
X = df.iloc[:, 0:45]
y = df.iloc[:, -1]
```

```python
print(df.dtypes)
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train.shape
```

```python
forward_feature_selection = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                     k_features= 6,
                                                     forward=True,
                                                     floating=False,
                                                     verbose=2,
                                                     scoring= "accuracy",
                                                     cv= 5).fit(X_train, y_train)
```

```python
forward_feature_selection.k_feature_idx_
```

```python
forward_feature_selection.k_feature_names_
```

```python
forward_feature_selection.k_score_
```

```python
pd.DataFrame.from_dict(forward_feature_selection.get_metric_dict()).T
```

```python
forward_feature_selection = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                     k_features= (1,11),
                                                     forward=True,
                                                     floating=False,
                                                     verbose=2,
                                                     scoring= "accuracy",
                                                     cv= 5).fit(X_train, y_train)
```

```python
forward_feature_selection.k_feature_names_
```

```python
forward_feature_selection.k_score_
```

```python
pd.DataFrame.from_dict(forward_feature_selection.get_metric_dict()).T
```

```python
import pickle

# Save the model
filename = "forward_feature_selection.sav"
pickle.dump(forward_feature_selection, open(filename, 'wb'))
```
